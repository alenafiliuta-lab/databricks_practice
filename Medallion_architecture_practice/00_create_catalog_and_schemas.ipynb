###Создание каталога и двух схем: bronze и infra

```python
%sql
CREATE CATALOG IF NOT EXISTS weather_stream;
USE CATALOG weather_stream;

CREATE SCHEMA IF NOT EXISTS weather_stream.bronze;
CREATE SCHEMA IF NOT EXISTS weather_stream.infra;

CREATE SCHEMA IF NOT EXISTS weather_stream.silver;
```

###Создание томов в схеме infra

```python
%sql
CREATE VOLUME IF NOT EXISTS weather_stream.infra.landing_volume;

CREATE VOLUME IF NOT EXISTS weather_stream.infra.autoloader_checkpoint;

CREATE VOLUME IF NOT EXISTS weather_stream.infra.streaming_checkpoints;

CREATE VOLUME IF NOT EXISTS weather_stream.infra.raw_data;
```
